```python
import matplotlib.pyplot as plt
%matplotlib inline

pd.plotting.scatter_matrix(df,alpha = 0.7, figsize=(20,20))
df.corr()
```

```python
def numeric_direction(value):
    if value == 'Up':
        return 1
    if value == 'Down':
        return 0

print("value of up is %d"%numeric_direction('Up'))
print("value of down %d"%numeric_direction('Down'))
```

```python
df['nDirection'] = df['Direction'].apply(numeric_direction)
df.info()
```

```python
df.head()
```

```python
df = df.drop('Direction', axis = 1)
```

```python
df
```

```python
worr = df.corr()
plt.matshow(worr.abs())
plt.colorbar()
plt.xticks(range(len(worr.columns)), worr.columns, rotation = 'vertical');
plt.yticks(range(len(worr.columns)), worr.columns)
```

```python
import statsmodels.formula.api as stf
model_1 = stf.logit('nDirection ~ Lag1 + Lag2 + Lag3 + Lag4 + Lag5 + Volume', df).fit()
```

```python
#print(dir(model_1))
model_1.summary()
```

```python
model_1.pvalues
```

```python
confusion_matrix = pd.DataFrame(model_1.pred_table())
print(confusion_matrix)
```

```python
confusion_matrix.columns=['Pred_class_0','Pred_Class_1']
confusion_matrix.index=['True_Class_0', 'True_Class_1']
print(confusion_matrix)
```

```python
sum_tmp = confusion_matrix.sum(axis =1)
print(sum_tmp, sum_tmp.shape)
```

```python
sum_tmp_1 = confusion_matrix.sum(axis = 0)
print(sum_tmp_1,sum_tmp_1.shape)
```

```python
confusion_matrix['Total'] = sum_tmp
confusion_matrix.loc['Total'] = sum_tmp_1
```

```python
print(confusion_matrix)
```
